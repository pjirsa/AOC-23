```python
def locate_first_numeric_digit(string):
    for char in string:
        if char.isdigit():
            return char
    return None
```

```python
def locate_last_numeric_digit(string):    
    last_digit = None
    for char in string:
        if char.isdigit():
            last_digit = char
    return last_digit
```

```python
lines = []
with open('Puzzle1Data.txt', 'r') as file:
    lines = file.readlines()

print('First numeric digit in each line:')
for line in lines:
    print(locate_first_numeric_digit(line))

print('\nLast numeric digit in each line:')
for line in lines:
    print(locate_last_numeric_digit(line))
```
